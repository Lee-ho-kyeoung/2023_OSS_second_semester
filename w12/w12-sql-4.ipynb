```python
### 참고 : 조회한 데이터를 판다스 프레임워크로 만들기

import pymysql

# !pip install panda
import pandas as pd
```

```python
sql = "SELECT * FROM countries" # 테이블 전체를 불러옴

try:
    with pymysql.connect(
        host='172.17.0.1', 
        port=3306, 
        user='root', 
        passwd='root', 
        db='nation', 
        charset='utf8'
    ) as db:
        cursor = db.cursor()
        cursor.execute(sql)
        rows = cursor.fetchall()
except pymysql.err.OperationalError as e:
    print(e)
```

```python
countries_df = pd.DataFrame(rows)
```
